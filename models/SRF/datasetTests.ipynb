```python
from convokit import Corpus, download
corpus = Corpus(filename=download("supreme-corpus"))
```

```python
print(f"Total Speakers: {len(corpus.get_speaker_ids())}")
print(f"Total Conversations: {len(corpus.get_conversation_ids())}")
print(f"Total Utterances: {len(corpus.get_utterance_ids())}")

for utt in corpus.iter_utterances():
    print(utt.speaker.id, ":", utt.text)
```
